1. Загрузите выборку Wine по адресу https://archive.ics.uci.edu/ml/machinelearning-databases/wine/wine.data

```python
import pandas as pd
import os
```

```python
data = pd.read_csv('wine.data', names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 
                                           'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
                                         'Proline' ] )
```

```python
data.head()
```

2. Подготовка данных (удалим не информативный столбец Class.

```python
X = data.drop(['Class'], axis = 1)
y = data.Class
```

4. Сделаем кросс-валидацию. Для этого сначала создаем генератор  KFold. Ему пеедаем что нужно 5 блоков, их нужно перемешать, а фиксированнныей параметр random_state = 42.

```python
import sklearn as sklearn
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state = 42)
kf.get_n_splits(data)
```

```python
for train_index, test_index in kf.split(data):
    print("TRAIN:", train_index, "TEST:", test_index)
```

5. Найдите точность классификации на кросс-валидации для метода
k ближайших соседей (sklearn.neighbors.KNeighborsClassifier), при
k от 1 до 50. При каком k получилось оптимальное качество? Чему
оно равно (число в интервале от 0 до 1)? Данные результаты и
будут ответами на вопросы 1 и 2.

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
import pandas as pd
cross = []
for i in range(1,51):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X, y)
    cross = cross_val_score(estimator=neigh, X = X, y = y, cv=kf, scoring='accuracy')
    m = cross.mean() 
    print('i',i,':',m)    
```

6. Произведите масштабирование признаков с помощью функции
sklearn.preprocessing.scale. Снова найдите оптимальное k на кроссвалидации

```python
X1 = sklearn.preprocessing.scale(X)
```
